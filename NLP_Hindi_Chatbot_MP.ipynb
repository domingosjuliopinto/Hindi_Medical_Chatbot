#Preprocessing

```python
#for preprocessing
from inltk.inltk import tokenize
from inltk.inltk import remove_foreign_languages
import nlu

def preprocessing(some_string):
 #tokenization
  tokenize_step = nltk.word_tokenize(some_string)

  #cleaning space removal
  space_removed_tokens = []

  for i in tokenize_step:
    if i[0]=="▁":
      new = i[1:]
      if new == "":
        continue
      space_removed_tokens.append(new)
    else:
      space_removed_tokens.append(i)

  #removing punctuations
  punc = [',',"!",".","।","|"]

  no_punc_tokens = []

  for i in space_removed_tokens:
    if i not in punc:
      no_punc_tokens.append(i)

  #removing stop words
  no_stop_words_token = []

  for w_token in no_punc_tokens:
    if w_token not in stop_words:
      no_stop_words_token.append(w_token)

  text1 = ""
  for word in no_stop_words_token:
    text1 += word
    text1 += " "

  return text1
```

```python
#For getting sickness name
def key_indices_sickness(x):
  y = x.partition("बीमारी: ")
  z = y[2]
  z = z.replace(".","")
  z = z.strip()
  return(z)
```

```python
#for getting the subtopic name
def key_indices_sub(x,type):
  z=""
  if type == 'v':
    y = x.partition("विवरण: ")
    z = y[1]
  if type == 'l':
    y = x.partition("लक्षण:  ")
    z = y[1]
  if type == 'k':
    y = x.partition("कारण: ")
    z = y[1]
  if type == 'e':
    y = x.partition("इलाज: ")
    z = y[1]
  if type == 'g':
    y = x.partition("घरेलू उपचार: ")
    z = y[1]
  z = z.replace(":","")
  z = z.strip()
  return z
```

# Setting up data set to our requirement

```python
f = open("Data.txt", "r", encoding='utf-8-sig')
corpus = f.read()

#corpus = corpus.replace(":",".")
corpus = corpus.replace("।",".")
corpus = corpus.replace("\n\n",". ")
corpus = corpus.replace("\n"," ")
corpus = corpus.replace("..",".")
corpus = corpus.replace("  "," ")
corpus = corpus.replace(". .",".")
hindi_sent_token = nltk.sent_tokenize(corpus)
```

```python
#first round: to get index of important points

k_b = 0
k_v = 0
k_l=0
k_k=0
k_e=0
k_g =0

indice_b = []
indice_v = []
indice_l=[]
indice_k=[]
indice_e=[]
indice_g=[]
countoftext = len(hindi_sent_token)
for i in range(countoftext):
  if "बीमारी:" in hindi_sent_token[i]:
    indice_b.append(i)
    k_b = k_b+1
  if "विवरण:" in hindi_sent_token[i]:
    indice_v.append(i)
    k_v = k_v+1
  if "लक्षण:" in hindi_sent_token[i]:
    indice_l.append(i)
    k_l = k_l+1
  if "कारण:" in hindi_sent_token[i]:
    indice_k.append(i)
    k_k = k_k+1
  if "इलाज:" in hindi_sent_token[i]:
    indice_e.append(i)
    k_e = k_e+1
  if "घरेलू उपचार:" in hindi_sent_token[i]:
    indice_g.append(i)
    k_g = k_g+1

k_b = k_b +1
indice_b.append(countoftext)
```

```python
#second round: to generate index dictionary
sickness_list = []
subtopic_list = ["विवरण","लक्षण","कारण","इलाज","घरेलू उपचार"]
main_dict = dict()
i_v = 0
i_l = 0
i_k = 0
i_e = 0
i_g = 0
for i_b in range(k_b-1):
  temp_dict = dict()
  temp_l = []
  temp_term = ""
  
  main_sick = key_indices_sickness(hindi_sent_token[indice_b[i_b]])
  sickness_list.append(main_sick)

  if indice_v[i_v] > indice_b[i_b] and indice_v[i_v] < indice_b[i_b+1]:
    temp_l.append(indice_v[i_v])
    temp_term = key_indices_sub(hindi_sent_token[indice_v[i_v]],'v')
    i_v = i_v + 1
  
  if indice_l[i_l] > indice_b[i_b] and indice_l[i_l] < indice_b[i_b+1]:
    temp_l.append(indice_l[i_l]-1)
    temp_dict[temp_term] = temp_l

    temp_l = []

    temp_l.append(indice_l[i_l])
    temp_term = key_indices_sub(hindi_sent_token[indice_l[i_l]],'l')
    i_l = i_l+1
  
  if indice_k[i_k] > indice_b[i_b] and indice_k[i_k] < indice_b[i_b+1]:
    temp_l.append(indice_k[i_k]-1)
    temp_dict[temp_term] = temp_l

    temp_l = []

    temp_l.append(indice_k[i_k])
    temp_term = key_indices_sub(hindi_sent_token[indice_k[i_k]],'k')
    i_k = i_k+1

  if indice_e[i_e] > indice_b[i_b] and indice_e[i_e] < indice_b[i_b+1]:
    temp_l.append(indice_e[i_e]-1)
    temp_dict[temp_term] = temp_l

    temp_l = []

    temp_l.append(indice_e[i_e])
    temp_term = key_indices_sub(hindi_sent_token[indice_e[i_e]],'e')
    i_e = i_e+1

  if indice_g[i_g] > indice_b[i_b] and indice_g[i_g] < indice_b[i_b+1]:
    temp_l.append(indice_g[i_g]-1)
    temp_dict[temp_term] = temp_l

    temp_l = []

    temp_l.append(indice_g[i_g])
    temp_term = key_indices_sub(hindi_sent_token[indice_g[i_g]],'g')
    i_g = i_g+1

  temp_l.append(indice_b[i_b+1]-1)
  temp_dict[temp_term] = temp_l  

  main_dict[main_sick] = temp_dict
```

#Other Required Functions

```python
#for choice checking
def choice_check(choice_str):
  if ("हाँ" in choice_str or "हां" in choice_str or "पता है" in choice_str) and ("ना" in choice_str or "नहीं" in choice_str or "इनकार" in choice_str or "अस्वीकार" in choice_str) :
    return 2
  
  elif "हाँ" in choice_str or "हां" in choice_str or "पता है" in choice_str:
    return 0
    
  elif "ना" in choice_str or "नहीं" in choice_str or "इनकार" in choice_str or "अस्वीकार" in choice_str:
    return 1
  
  else :
    return 3
```

```python
#query processing
def query_processing(user_query):
  preprocessed_user_query = preprocessing(user_query)
  list_preprocessed_user_query = preprocessed_user_query.split(" ")
  list_preprocessed_user_query.remove('') 
  return list_preprocessed_user_query 
```

```python
#finding sickness from symptoms
def symptoms_to_sickness(lopq):
  probability_count = 0
  sick_name = ""
  sick_flag = []
  for m in sickness_list:
    try:
      min_span = main_dict[m]['लक्षण'][0]
      max_span = main_dict[m]['लक्षण'][1]
      check_hst = hindi_sent_token[min_span:max_span+1]
      temp_pc = 0
      for n in lopq:
        for o in check_hst:
          if n in o:
            temp_pc = temp_pc + 1
            break
      if temp_pc > probability_count:
        probability_count = temp_pc
        sick_name = m
        sick_flag =[m]
      if temp_pc == probability_count:
        if m not in sick_flag:
          sick_name = sick_name+" या "+m
    except KeyError:
      pass
  guess= "हमारी जानकारी के आधार आपको "+sick_name+ " है |"
  return guess
```

```python
def knowledge_output(key1,key2,key_emergency,user_query):
  #final knowledge output
  if (key1 != "" and key2 != ""):
    try:
      min_span = main_dict[key1][key2][0]
      max_span = main_dict[key1][key2][1]
      shortened_hst = hindi_sent_token[min_span:max_span+1]
      print(colored("चैटबोट:", "green"))
      for sentence in shortened_hst:
        print(colored(sentence, "green"))
    except KeyError:
      print(colored("चैटबोट: हमें खेद है कि इस समय हमारे पास वह ज्ञान नहीं है", "green"))
  elif (key1 != "" and key2 == ""):
    try:
      min_span = main_dict[key1]['विवरण'][0]
      max_span = main_dict[key_emergency]['विवरण'][0] - 2
      if "क्या होता है" in user_query:
        try:
          min_span = main_dict[key1]['लक्षण'][0]
          max_span = main_dict[key1]['लक्षण'][1]
        except KeyError:
          min_span = main_dict[key1]['विवरण'][0]
          max_span = main_dict[key1]['विवरण'][1]
      if "क्यों होता है" in user_query or "कैसे होता है" in user_query:
        try:
          min_span = main_dict[key1]['कारण'][0]
          max_span = main_dict[key1]['कारण'][1]
        except KeyError:
          min_span = main_dict[key1]['विवरण'][0]
          max_span = main_dict[key1]['विवरण'][1]
      if "कैसे ठीक" in user_query or "कैसे मुक्त" in user_query or "क्या करूँ" in user_query or "क्या करे" in user_query:
        min_span = main_dict[key1]['इलाज'][0]
        max_span = main_dict[key1]['इलाज'][1]
      shortened_hst = hindi_sent_token[min_span:max_span+1]
      print(colored("चैटबोट:", "green"))
      for sentence in shortened_hst:
        print(colored(sentence, "green"))
    except KeyError:
      print(colored("चैटबोट: हमें खेद है कि इस समय हमारे पास वह ज्ञान नहीं है", "green"))
  elif (key1 == "" and key2 != ""):
    print(colored("चैटबोट: आप किस बीमारी का "+key2+" चाहते हैं\n", "green"))
    user_reply = input("मरीज़: ")
    processed_ur = query_processing(user_reply)
    for i in processed_ur:
      for j in range(len(sickness_list)):
        if i in sickness_list[j] and key1 == "":
          key1 = sickness_list[j]
          key_emergency = sickness_list[j+1]
          break
    try:
      min_span = main_dict[key1][key2][0]
      max_span = main_dict[key1][key2][1]
      shortened_hst = hindi_sent_token[min_span:max_span+1]
      print(colored("चैटबोट:", "green"), end=" ")
      for sentence in shortened_hst:
        print(colored(sentence, "green"))
    except KeyError:
      print(colored("हमें खेद है कि इस समय हमारे पास वह ज्ञान नहीं है", "green"))
  else:
    print(colored("हमें खेद है कि इस समय हमारे पास वह ज्ञान नहीं है", "green"))
```

```python
#getting sickness knowledge
def sickness_knowledge(user_query,lopq):
  #Getting keywords to reduce dataset size
  key1 = ""
  key2 = ""
  key_emergency = ""
  sk_flag=1
  for i in lopq:
    for j in range(len(sickness_list)):
      if i in sickness_list[j] and key1 == "":
        key1 = sickness_list[j]
        key_emergency = sickness_list[j+1]
        break
    for k in subtopic_list :
      if i == k and key2 == "":
        key2 = i
        break
  knowledge_output(key1,key2,key_emergency,user_query)
  while sk_flag != 0:
    if(key1 != ""):
      print(colored("\nचैटबोट: क्या आप "+key1+" के बारे में कुछ और जानना चाहते हैं?\n", "green"))
      knowmore = input("मरीज़: ")
      returned_km = choice_check(knowmore)
      if returned_km == 0:
        print(colored("चैटबोट: आप "+key1+" के बारे में और क्या जानना चाहते हैं\n", "green"))
        user_reply = input("मरीज़: ")
        processed_ur = query_processing(user_reply)
        for i in processed_ur:
          for k in subtopic_list :
            if i == k:
              key2 = i
              break
        knowledge_output(key1,key2,key_emergency,user_query)
      elif returned_km == 1:
        sk_flag=0
      elif returned_km == 2:
        print(colored("चैटबोट: आप एक ही समय में 'हां' और 'नहीं' नहीं कह सकते", "green"))
        print(colored("चैटबोट: फिर से पूछ रहा हूँ", "green"))
      else:
        print(colored("चैटबोट: आपने न तो 'हां' कहा और न ही 'नहीं' कहा।", "green"))
        print(colored("चैटबोट: उचित सम्मान के साथ हम मान रहे हैं कि आप और जानना नहीं चाहते", "green"))
        sk_flag=0
    else:
      sk_flag=0
```
